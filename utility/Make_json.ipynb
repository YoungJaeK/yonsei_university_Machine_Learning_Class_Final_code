## 4Class json maker

```python
# 4 classes
# make a json file 
import glob
import json
import os

ROOT_PATH = '/disk1/yjkim/data/endo/intestine/ver8'
SAVE_ROOT_PATH = '/disk1/yjkim/json/intestine'
NAME_OF_FOLDER = ["normal","lgd",'hgd','cancer']
OUTPUT_PATH = os.path.join(SAVE_ROOT_PATH,"c4json")

TRAIN_PATH_LIST = []
VAL_PATH_LIST = []
TEST_PATH_LIST = []

TRAIN_LABEL_LIST = []
VAL_LABEL_LIST = []
TEST_LABEL_LIST = []

for i in range(len(NAME_OF_FOLDER)):
    # read data path
    PATH_LIST = glob.glob(os.path.join(ROOT_PATH,NAME_OF_FOLDER[i],'*.*'), recursive=True)
    
    POINT_TEST = int(len(PATH_LIST)*0.8)
    POINT_VAL = int(POINT_TEST*0.8)
    
    TRAIN_PATH_LIST += PATH_LIST[:POINT_VAL]
    VAL_PATH_LIST += PATH_LIST[POINT_VAL:POINT_TEST]
    TEST_PATH_LIST += PATH_LIST[POINT_TEST:]
    
    # set label
    LABEL_LIST = []
    for j in range(len(PATH_LIST)):
        LABEL = i
        LABEL_LIST.append(LABEL)
        
    TRAIN_LABEL_LIST += LABEL_LIST[:POINT_VAL]
    VAL_LABEL_LIST += LABEL_LIST[POINT_VAL:POINT_TEST]
    TEST_LABEL_LIST += LABEL_LIST[POINT_TEST:]
    
#     assert len(TEMP_PATHES) == len(LABEL_LIST)

# Save to file
# train
with open(os.path.join(OUTPUT_PATH, 'TRAIN_images.json'), 'w') as j:
    json.dump(TRAIN_PATH_LIST, j)
with open(os.path.join(OUTPUT_PATH, 'TRAIN_labels.json'), 'w') as j:
    json.dump(TRAIN_LABEL_LIST, j)
# val
with open(os.path.join(OUTPUT_PATH, 'VAL_images.json'), 'w') as j:
    json.dump(VAL_PATH_LIST, j)
with open(os.path.join(OUTPUT_PATH, 'VAL_labels.json'), 'w') as j:
    json.dump(VAL_LABEL_LIST, j)
# test
with open(os.path.join(OUTPUT_PATH, 'TEST_images.json'), 'w') as j:
    json.dump(TEST_PATH_LIST, j)
with open(os.path.join(OUTPUT_PATH, 'TEST_labels.json'), 'w') as j:
    json.dump(TEST_LABEL_LIST, j)
```

## HCC Json file

```python
import os
import glob
import json
# Making a json file
from yjlib.od.utils import create_data_lists
import xml.etree.ElementTree as ET
from yjlib.od.utils import parse_annotation

HCC_PATH = "/disk1/yjkim/data/hcc1"
OUTPUT_PATH = os.path.join(HCC_PATH,"jsonpath")
# paths // 폴더 안의 파일들 path list 불러오기
dcm_paths = glob.glob(os.path.join(HCC_PATH,"dcmImages","*.dcm"))


def hcc_parse_annotation(annotation_path):
    tree = ET.parse(annotation_path)
    root = tree.getroot()
    
    boxes = list()
    labels = list()
    label_map = {"background":0,"hcc":1}
    
    for object in root.iter('object'):
        label = "hcc"
        if label not in label_map:
            continue
        bbox = object.find('bndbox')
        xmin = int(bbox.find('xmin').text) - 1
        ymin = int(bbox.find('ymin').text) - 1
        xmax = int(bbox.find('xmax').text) - 1
        ymax = int(bbox.find('ymax').text) - 1

        boxes.append([xmin, ymin, xmax, ymax])
        labels.append(label_map[label])
    
    return {'boxes': boxes, 'labels': labels}

def create_hcc_data_lists(paths,  output_folder):
    paths = os.path.abspath(paths)
    #voc12_path = os.path.abspath(voc12_path)

    train_images = list()
    train_objects = list()

    # Training data
    for path in dcm_paths[:-148]:
        file = path.split(sep="/")[-1] # 맨마지막
        idx = file.replace(".dcm","") # 확장자 제거
        objects = hcc_parse_annotation(os.path.join(HCC_PATH, 'Annotations', idx + '.xml'))
        train_objects.append(objects)
        train_images.append(os.path.join(HCC_PATH, 'dcmImages', idx + '.dcm'))
            
    assert len(train_objects) == len(train_images)

    
    
    # Save to file
    with open(os.path.join(output_folder, 'TRAIN_images.json'), 'w') as j:
        json.dump(train_images, j)
    with open(os.path.join(output_folder, 'TRAIN_objects.json'), 'w') as j:
        json.dump(train_objects, j)
    with open(os.path.join(output_folder, 'label_map.json'), 'w') as j:
        json.dump(label_map, j)  # save label map too

    print('\nThere are %d training images. Files have been saved to %s.' % (
        len(train_images), os.path.abspath(output_folder)))

    # Validation data
    test_images = list()
    test_objects = list()
    
    for path in dcm_paths[-148:]:
        file = path.split(sep="/")[-1] # 맨마지막
        idx = file.replace(".dcm","") # 확장자 제거
        objects = hcc_parse_annotation(os.path.join(HCC_PATH, 'Annotations', idx + '.xml'))
        test_objects.append(objects)
        test_images.append(os.path.join(HCC_PATH, 'dcmImages', idx + '.dcm'))
            
    assert len(test_objects) == len(test_images)

    # Save to file
    with open(os.path.join(output_folder, 'TEST_images.json'), 'w') as j:
        json.dump(test_images, j)
    with open(os.path.join(output_folder, 'TEST_objects.json'), 'w') as j:
        json.dump(test_objects, j)

    print('\nThere are %d validation images. Files have been saved to %s.' % (
        len(test_images),os.path.abspath(output_folder)))
```
